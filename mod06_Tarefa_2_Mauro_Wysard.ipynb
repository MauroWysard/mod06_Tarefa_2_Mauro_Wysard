```python
#TENTATIVA DE ORGANIZAR OS HISTOGRAMAS EM LINHAS E COLUNAS. EXISTE UMA FORMA DE FAZER ISSO?

sns.displot(x='IDADEMAE', data=df, row=0, col=0, alpha=0.5, bins=10, element='step', kde=True)
sns.displot(x='IDADEMAE', data=df, row=0, col=1, alpha=0.5, bins=20, element='step', kde=True)
sns.displot(x='IDADEMAE', data=df, row=1, col=0, alpha=0.5, bins=30, element='step', kde=True)
sns.displot(x='IDADEMAE', data=df, row=1, col=1, alpha=0.5, bins=40, element='step', kde=True)
sns.displot(x='IDADEMAE', data=df, row=2, col=0, alpha=0.5, bins=50, element='step', kde=True)
sns.displot(x='IDADEMAE', data=df, row=2, col=1, alpha=0.5, bins=100, element='step', kde=True)
```

```python
#TENTATIVA DE ORGANIZAR OS HISTOGRAMAS EM LINHAS E COLUNAS. EXISTE UMA FORMA DE FAZER ISSO?

plt.rc('figure', figsize=(20,15))
fig, axes = plt.subplots(3,2)

sns.displot(ax = axes[0, 0], x='IDADEMAE', data=df, alpha=0.5, bins=10, element='step', kde=True)
sns.displot(ax = axes[0, 1], x='IDADEMAE', data=df, alpha=0.5, bins=20, element='step', kde=True)
sns.displot(ax = axes[1, 0], x='IDADEMAE', data=df, alpha=0.5, bins=30, element='step', kde=True)
sns.displot(ax = axes[1, 1], x='IDADEMAE', data=df, alpha=0.5, bins=40, element='step', kde=True)
sns.displot(ax = axes[2, 0], x='IDADEMAE', data=df, alpha=0.5, bins=50, element='step', kde=True)
sns.displot(ax = axes[2, 1], x='IDADEMAE', data=df, alpha=0.5, bins=100, element='step', kde=True)
```
